```python
#  Assignemnt 3 from Jeff Heaton's 2020 "T81-558: Applications of Deep Neural Networks"
# v1

# Simple dataframe
import os
import pandas as pd
import numpy as np

# import datetime utilities
from datetime import datetime

# read input file from web
df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/datasets/series-31.csv")

#df['time'] = pd.to_datetime(df['time'], errors='coerce')


# Parse cvs date and time into python datetime
col_datetime = df['time']
col_date_only = []
col_time_only = []

for xxx in col_datetime:
    dt = datetime.strptime(xxx, "%m-%d-%Y %H:%M")
    col_date_only.append(dt.date())
    col_time_only.append(dt.time())

# make new df with date only and time only fields
df2 = df.copy()
df2["date_only"] = col_date_only
df2["time_only"] = col_time_only

# Make list of unique dates
u_date = np.unique(col_date_only)



#setup stats based each unique date
vstats = []

for x in u_date:
    

    df3 = df2.loc[df2['date_only']==x]
    
    #sort on time
    df4 = df3.sort_values(by='time_only')

    # build stats
    x1 = list(df4['value'])
    vstats.append({
        'date' : x,
        'starting': x1[0],
        'max': df4['value'].max(),
        'min': df4['value'].min(),
        'ending': x1[-1]
    })
    
# Build new data frame with stats
df5 = pd.DataFrame(vstats)
display(df5)

 
```
