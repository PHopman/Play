```python
import os
import pandas as pd
from scipy.stats import zscore
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import pandas as pd
import io
import requests
import numpy as np
from sklearn import metrics

# You must also identify your source file.  (modify for your local setup)
# file='/content/drive/My Drive/Colab Notebooks/assignment_yourname_class3.ipynb'  # Google CoLab
# file='C:\\Users\\jeffh\\projects\\t81_558_deep_learning\\assignments\\assignment_yourname_class3.ipynb'  # Windows
file='/Users/jheaton/projects/t81_558_deep_learning/assignments/assignment_yourname_class3.ipynb'  # Mac/Linux

# Begin assignment
df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/crx.csv",na_values=['?'])
```

```python
df['a9'] = pd.Series(df['a9'].values).astype('category').cat.codes.values+1
df['a10'] = pd.Series(df['a10'].values).astype('category').cat.codes.values+1
df['a11'] = pd.Series(df['a11'].values).astype('category').cat.codes.values+1
df['a12'] = pd.Series(df['a12'].values).astype('category').cat.codes.values+1
df['a13'] = pd.Series(df['a13'].values).astype('category').cat.codes.values+1
df = df.fillna(df.median())
# Pandas to Numpy
x = df[['s3', 'a8', 'a9', 'a10', 'a11', 'a12', 'a13', 'a15']].values
y = df['a14'].values # regression
model = Sequential()
model.add(Dense(25, input_dim=x.shape[1], activation='relu')) # Hidden 1
model.add(Dense(10, activation='relu')) # Hidden 2
model.add(Dense(1)) # Output
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x,y,verbose=2,epochs=100)
```

```python
pred = model.predict(x)
print("Shape: {}".format(pred.shape))
print(pred)
```
